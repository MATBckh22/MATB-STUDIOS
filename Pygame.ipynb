# Working with Pygame

Pygame module offers functions for creating GUI programs which can show images and colors.

```python
import pygame, sys
from pygame.locals import * #loads whole module

pygame.init()
height = 300
width = 400
DISPLAYSURF = pygame.display.set_mode((width, height))
pygame.display.set_caption('Hello World!')
# Main game loop
while True:
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            sys.exit()
    pygame.display.update()
```

- set height and width using `height` and `width`, in this case it's 400 pixels wide and 300 pixels tall
- `set_caption` sets the title of the window
- `DISPLAYSURF` returns the window
- the while loop runs forever until the user exits the window, pygame window and sys quits

```python
import random
import pygame, sys
from pygame.locals import *
import time

pygame.init()

# Initialize Pygame
pygame.init()

# Set up the display
WINDOW_WIDTH = 400
WINDOW_HEIGHT = 600
window = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("Dodge My Bullets")

# Define colors
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# Player character
player = pygame.Rect(200, 500, 30, 30)

# List to store falling obstacles
obstacles = []

# Variables for obstacle generation
OBSTACLE_SPEED = 5
OBSTACLE_SPAWN_RATE = 25  # Adjust to change frequency of obstacles
score = 0

running = True
Clock = pygame.time.Clock()
# Main game loop
while True:
    window.fill(WHITE)  # Clear the screen

    # Event handling
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            sys.exit()

    # Player movement
    keys = pygame.key.get_pressed()
    if keys[K_LEFT]:
        player.x -= 5
    if keys[K_RIGHT]:
        player.x += 5

    # Generate falling obstacles
    if random.randint(1, OBSTACLE_SPAWN_RATE) == 1:
        obstacle = pygame.Rect(random.randint(0, WINDOW_WIDTH - 20), 0, 20, 20)
        obstacles.append(obstacle)

    # Move and draw obstacles
    for obstacle in obstacles:
        obstacle.y += OBSTACLE_SPEED
        pygame.draw.rect(window, RED, obstacle)

        # Check for collision with player
        if obstacle.colliderect(player):
            print("Game Over! Your score:", score)
            time.sleep(2)
            pygame.quit()
            sys.exit()

        # Remove obstacles when they go off-screen
        if obstacle.y > WINDOW_HEIGHT:
            obstacles.remove(obstacle)
            score += 1
    # Draw player character
    pygame.draw.rect(window, RED, player)

    # Display score
    font = pygame.font.Font(None, 36)
    score_text = font.render("Score: " + str(score), True, RED)
    window.blit(score_text, (10, 10))

    pygame.display.update()
    Clock.tick(30)  # Limit frame rate to 30 FPS
```

```python
import random
import pygame, sys
from pygame.locals import *
import time

pygame.init()

# Initialize Pygame
pygame.init()

# Set up the display
WINDOW_WIDTH = 400
WINDOW_HEIGHT = 600
window = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("Harder Flappy Bird")

# Define colors
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# Player character
player = pygame.Rect(200, 500, 30, 30)

# List to store falling obstacles
obstacles = []

# Variables for obstacle generation
OBSTACLE_SPEED = 5
OBSTACLE_SPAWN_RATE = 25  # Adjust to change frequency of obstacles
score = 0

running = True
Clock = pygame.time.Clock()
# Main game loop
while True:
    window.fill(WHITE)  # Clear the screen

    # Event handling
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            sys.exit()

    # Player movement
    keys = pygame.key.get_pressed()
    gravity = 1/2*9.8*1**2
    if keys[K_UP]:
        player.y -= 10

    player.y += gravity
    # Generate falling obstacles
    if random.randint(1, OBSTACLE_SPAWN_RATE) == 1:
        # Set the initial position of the obstacle outside the visible area of the window
        side = random.choice(["left", "right"])
        if side == "left":
            obstacle = pygame.Rect(-20, random.randint(0, WINDOW_HEIGHT - 20), 20, 20)
        else:
            obstacle = pygame.Rect(WINDOW_WIDTH, random.randint(0, WINDOW_HEIGHT - 20), 20, 20)
        obstacles.append(obstacle)

    # Move and draw obstacles
    for obstacle in obstacles:
        obstacle.x += OBSTACLE_SPEED
        pygame.draw.rect(window, RED, obstacle)

        # Check for collision with player
        if obstacle.colliderect(player) or player.y > WINDOW_HEIGHT:
            print("Game Over! Your score:", score)
            time.sleep(2)
            pygame.quit()
            sys.exit()

        # Remove obstacles when they go off-screen
        if obstacle.y > WINDOW_WIDTH:
            obstacles.remove(obstacle)
            score += 1
    # Draw player character
    pygame.draw.rect(window, RED, player)

    # Display score
    font = pygame.font.Font(None, 36)
    score_text = font.render("Score: " + str(score), True, RED)
    window.blit(score_text, (10, 10))

    pygame.display.update()
    Clock.tick(30)  # Limit frame rate to 30 FPS
```

```python
import pygame
import sys

pygame.init()

width, height = 800,600
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Draw rectangle")

WHITE = (255, 255, 255)
BLACK = (0, 0 ,0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)


rect_width = 50
rect_height = 50
rect_x = (width - rect_width) // 2
rect_y = (height - rect_height) // 2

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    screen.fill(WHITE)
    pygame.draw.circle(screen, RED, (50, 20), 10)
    pygame.draw.rect(screen, GREEN, (rect_x, rect_y, rect_width, rect_height))
    pygame.draw.line(screen, BLUE, (20, 8), (50, 20), 2)
    pygame.display.flip()

pygame.quit()
sys.quit()
```
