```python
#prints a random number between 1-7 for 10 and 100 iterations
few_rolls = np.random.randint(1,7,size=10)
many_rolls = np.random.randint(1,7,size=100)
print(few_rolls)
print(many_rolls)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#plotting histogram plots

#plot the distriution of few_rolls(10 sets of random numbers between 1-7)
few_counts = np.histogram(few_rolls, bins=np.arrange(0.5, 7.5))[0] #[0] only reports 

#plot the distriution of few_rolls(100 sets of random numbers between 1-7)
many_counts= np.histogram(many_rolls, bins=np.arrange(0.5, 7.5))[0]
```
